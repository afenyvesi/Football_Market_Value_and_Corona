# Rearranging the data

We want to have 5 rows for each player, each row representing a season of performance and market value.  

The observed seasons are: 2016/17, 2017/18, 2018/19, 2019/20 before the shutdown, 2019/20 after the shutdown.  

Each player's data is stored in dictionaries, we will extract them from there.

```python
def create_model_df(players):
    marketvaluedates = [['Apr2020'], ['May2019', 'Jun2019', 'Jul2019', 'Aug2019'], ['May2018', 'Jun2018', 'Jul2018', 'Aug2018'] , ['May2017', 'Jun2017', 'Jul2017', 'Aug2017']]
    outputdf = []
    for player in players:
        yearssince = 0
        for season in ['19/20', '18/19', '17/18', '16/17']:
            playerrow = {}
            playerrow['season'] = season
            playerrow['name'] = player['name']
            whichvalue = 0
            for valuedate in player['marketvaluedata']['date']:
                if valuedate[:3] + valuedate[-4:] in marketvaluedates[yearssince]:
                    playerrow['marketvalue'] = float(player['marketvaluedata']['market_value'][whichvalue])
                    try:
                        playerrow['prev_marketvalue'] = float(player['marketvaluedata']['market_value'][whichvalue - 1])
                    except:
                        playerrow['prev_marketvalue'] = None
                else:
                    whichvalue += 1
            try:
                playerrow['log_marketvalue'] = math.log(playerrow['marketvalue'])
            except:
                playerrow['log_marketvalue'] = None
            try:
                playerrow['log_prev_marketvalue'] = math.log(playerrow['prev_marketvalue'])
            except:
                playerrow['log_prev_marketvalue'] = None
            try:
                playerrow['height_cm'] = float(player['height'].replace('m','').replace(' ','').replace(',','.'))*100
            except:
                playerrow['height_cm'] = None
            playerrow['age'] = int(player['dateofbirth'][-3:-1]) - yearssince
            playerrow['position'] = player['genposition']
            playerrow['foot'] = player['foot']
            try:
                performancesum = performance_sum(player).loc[season]
                for i in range(len(performancesum)):
                    playerrow[performancesum.index[i]] = performancesum.values[i]
            except:
                pass
            yearssince += 1
            outputdf.append(playerrow)
    return pd.DataFrame(outputdf)
```

```python
def performance_sum(player):
    proba_df = pd.DataFrame(player['performancestats'])

    for column in ['in_squad', 'appearances', 'goals', 'assists', 'owngoals', 'sub_on', 'sub_off', 'yellow', 'secondyellow', 'red', 'penalty']:
        proba_df[column] = [int(x) if x != '-' else None for x in proba_df[column]]
    
    proba_df['pointspermatch'] = [float(x.replace(',','.')) if x != '-' else None for x in proba_df['pointspermatch']]
    proba_df['minutes_played'] = [float(x[:-1].replace('.','').replace(',','.')) if x != '-' else None for x in proba_df['minutes_played']]

    return pd.pivot_table(proba_df,
               values = ['minutes_played', 'pointspermatch', 'in_squad', 'appearances', 'goals', 'assists', 'owngoals', 'sub_on', 'sub_off', 'yellow', 'secondyellow', 'red', 'penalty'],
               index = ['season'],
               aggfunc = {'minutes_played' : np.sum,
                         'pointspermatch' : np.mean,
                         'in_squad' : np.sum,
                         'appearances' : np.sum,
                         'goals' : np.sum,
                         'assists' : np.sum,
                         'owngoals' : np.sum,
                         'sub_on' : np.sum,
                         'sub_off' : np.sum,
                         'yellow' : np.sum,
                         'secondyellow' : np.sum,
                         'red' : np.sum,
                         'penalty' : np.sum})#.loc[('13/14')]
```

# Transfer data export

```python
marketvaluedates = [['Apr2020'], ['May2019', 'Jun2019', 'Jul2019', 'Aug2019'], ['May2018', 'Jun2018', 'Jul2018', 'Aug2018'] , ['May2017', 'Jun2017', 'Jul2017', 'Aug2017']]
```

```python
transfers = []

for player in players:
    for transfer in player['playertransfers']:
        transferrow = {}
        try:
            transferrow['player_name'] = player['name']
            transferrow['genposition'] = player['genposition']
            transferrow['specposition'] = player['specposition']
            transferrow['dateofbirth'] = player['dateofbirth']
            transferrow['nationality'] = player['nationality']
            transferrow['height'] = player['height']
            transferrow['foot'] = player['foot']
            transferrow['playerurl'] = player['playerurl']
        except:
            continue
        try:
            transferrow['season'] = transfer['season']
            transferrow['date'] = transfer['date']
            transferrow['from_club'] = transfer['from_club']
            transferrow['to_club'] = transfer['to_club']
            transferrow['value'] = transfer['value']
            transferrow['price'] = transfer['price']
        except:
            continue
        transfers.append(transferrow)
```

```python
def parse_value(valuestring):
    if valuestring[-3:] == 'Th.':
        try:
            return float(valuestring.replace('£','').replace('Th.','').replace('Â','')) * 1000
        except:
            return None
    elif valuestring[-1:] == 'm':
        try:
            return float(valuestring.replace('£','').replace('m','').replace('Â','')) * 1000000
        except:
            return None
    else:
        return None 
```

```python
transfersdf = pd.DataFrame(transfers)
transfersdf['parsedvalue'] = transfersdf['value'].apply(lambda x : parse_value(x))
transfersdf['parsedprice'] = transfersdf['price'].apply(lambda x : parse_value(x))
transfersdf
```

```python
transfersdf.to_csv('transferdata.csv', sep = ';', encoding = 'utf-8')
```
